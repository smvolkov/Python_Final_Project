```python
def bisection(f, lhs, rhs, eps=1e-6):

  mid = (lhs + rhs) / 2

  while abs(lhs - rhs) > eps:
    mid = (lhs + rhs) / 2

    if f(lhs) * f(mid) > 0:
      lhs = mid
    else:
      rhs = mid
      
  return mid
```
